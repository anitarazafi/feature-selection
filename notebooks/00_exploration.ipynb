## Data exploration

```python
import pandas as pd
from src.preprocessing.load_dataset import load_dataset
from src.preprocessing.normalize_missing_values import normalize_missing_values
```

```python
ds = "ibtracs.last3years"
df, cfg = load_dataset(ds)
```

```python
# drop duplicated header/metadata row
df = df.iloc[1:].reset_index(drop=True)
```

```python
df_copy = normalize_missing_values(df)
```

```python
missing_counts = df_copy.isna().sum()
missing_percent = (df_copy.isna().mean() * 100)
missing_summary = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_percent": missing_percent
}).sort_values("missing_count")
pd.set_option("display.max_rows", None)
missing_summary
```
